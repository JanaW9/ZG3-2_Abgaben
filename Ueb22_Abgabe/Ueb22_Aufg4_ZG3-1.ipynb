```python
class Verkehrsmittel:  #Klasse Verkehrsmittel ist die Oberklasse
    def __init__(self,preis,anzahlr,farbe): #Konstruktor, initalisiert die Klassenattribute.
        self.preis=preis
        self.anzahlreifen=anzahlr
        self.farbe=farbe
  
    def __str__(self): # __str__() Methode gibt ein String zurück wenn man die Referenzvariable eines Objekts in print schreibt.Das ist eine Built-In Methode.
        erg='Verkehrsmittel das: '+str(self.preis)+' € kostet. Anzahl Reifen: '+str(self.anzahlreifen)+' Farbe: '+self.farbe
        return erg #string  als Rückgabewert
class Fahrrad(Verkehrsmittel): #Unterklasse erkennt man an der Syntax: Klasse(Oberklasse). Ist ein Verweis auf die Oberklass
                                #die Unterklasse erbt Attribute und Methoden der Oberklasse
    def __init__(self,rahmennr,preis,anzahlr,farbe,marke,kosename): #Konstruktor der Oberklasse, Rahmennr, Marke neue zusätzliche Attribute
        super().__init__(preis,anzahlr,farbe) #Konstruktoraufruf der Oberklasse mit super=>ist ein Verweis auf die Oberklasse
        self.fahrradid=rahmennr 
        self.marke=marke
        self.kosename=kosename #Kosename Attribut hinzugefügt
    def __eq__(self, other):#__eq__()(equal)- Methode prüft die inhaltliche Gleichheit von zwei Objekten
        return self.kosename==other.kosename #gibt Boolschen Wert (true/false zurück)
    def __lt__(self, other): #__lt__() (lower than)- Methode, prüft ob das self Objekt kleiner als das other Objekt ist.
        return self.farbe < other.farbe #gibt true/false zurück
    def __str__(self): 
        verk=super().__str__()
        erg='Fahrrad von der Marke '+self.marke+ ' mit Rahmennr: '+str(self.fahrradid)+ ' der Kosename lautet: '+self.kosename
        return verk+'\n'+erg 
v1=Verkehrsmittel(120,4,'grün') #Objekt der Oberklasse
f1=Fahrrad(123456,20,2,'blau','Radon','Erna') #Objekt der Unterklasse
f2=Fahrrad(123457,59,2,'rot','BTWIN','Apfel')
f3=Fahrrad(123458,120,2,'schwarz','Pegasus','Erna')
print('\n c)Ausgabe der Objekte \n')
print(v1)
print(f1)
print(f2)
print(f3)
print('\n f)_______Unsortierte Ausgabe der Fahrradobjekte________ \n')
Fahrräder = [f1,f2,f3]
for f in Fahrräder:
    print(f)
print('\n f)__Sortierte Ausgabe der Fahrradobjekte_________ \n')
Fahrräder.sort() #die Objekte werden sortiert nach der Methode__lt__(Die Farbe wird alphabetisch, der größe aufsteigend sortiert.)
for b in Fahrräder:
    print(b) 

print('\n e) Prüfung der Objekt-Gleichheit \n')
print(f1==f3) #die Objekte haben den gleichen Kosenamen. Die Objekte sind für das Attritbut Kosenamen wertgleich, aber nicht identisch.
print('\n e) Prüfung der Objekt-Identität \n')
print(f1 is f3) 
```
